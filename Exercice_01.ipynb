## Bibliothèques

```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
```

## Question 1 


```python
#  Charger une image 
image = cv2.imread("image1.jpg")
```

## Question 2

```python
# 2. Histogramme
colors = ('b', 'g', 'r')
for i, color in enumerate(colors):
    hist = cv2.calcHist([image], [i], None, [256], [0, 256])
    plt.plot(hist, color=color)
plt.xlabel('Intensité des pixels')
plt.ylabel('Nombre de pixels')
plt.grid()
plt.title('Histogramme en couleur')
plt.show()

```

## Question 3

```python
# 3. Bruit Gaussien
mean = 0
std_dev = 7
noise = np.random.normal(mean, std_dev, image.shape).astype(np.uint8)
```

## Question 4

```python
# 4. Créer l'image bruitée et l'afficher
noisy_image = cv2.add(image, noise)
cv2.imshow('Image bruitée', noisy_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Question 5

```python
# 5. Charger une image et la convertir en niveaux de gris
gray_image = cv2.imread("votre_image.jpg", cv2.IMREAD_GRAYSCALE)
```

## Question 6

```python
# 6. Afficher les gradients
sobelx = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=5)
sobely = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=5)
gradient_magnitude = np.sqrt(sobelx**2 + sobely**2)

plt.imshow(gradient_magnitude, cmap='gray')
plt.axis('off')
plt.title('Gradients')
plt.show()

```

## Question 7

```python
# 7. Détecter les contours
edges = cv2.Canny(gray_image, 100, 200)
```
